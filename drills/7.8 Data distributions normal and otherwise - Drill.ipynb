### 1. Generate two normally-distributed variables, one with a mean of 5 and standard deviation of 0.5, and the other with a mean of 10 and standard deviation of 1.


```python
pop1 = np.random.normal(5, .5, 1000)
pop2 = np.random.normal(10, 1, 1000)
```

### 2. Add them together to create a third variable.


```python
pop3 = pop1 + pop2
```

### 3. Graph the third variable using a histogram.


```python
plt.figure(figsize=(8,4))
plt.hist(pop3, alpha=.6, color='g' )
plt.show()
```

### 4. Compute the mean and standard deviation and plot them as vertical lines on the histogram.


```python
plt.figure(figsize=(8,4))
plt.hist(pop3, alpha=.6, color='g', label='Histogram of combined populations')
plt.axvline(pop3.mean(), color='g', linestyle='dashed', linewidth=1, label='Mean')
plt.axvline(pop3.mean() + pop3.std(), color='m', linestyle='dashed', linewidth=1, label='upper end of standard deviation')
plt.axvline(pop3.mean() - pop3.std(), color='b', linestyle='dashed', linewidth=1, label='lower end of standard deviation')
plt.legend(loc='upper right')
plt.title('combined populations')
plt.xlabel('values')
plt.ylabel('observations')

plt.show()
```

### 5. Eval1uate the descriptive statistics against the data.

```python
print('for pop3, the mean is {} and standard deviation is {}'.format(pop3.mean(), pop3.std()))
```

The descriptive statistics do a good job of showing where the bulk of hte data is, though a fair amount is outside the first standard deviation.
